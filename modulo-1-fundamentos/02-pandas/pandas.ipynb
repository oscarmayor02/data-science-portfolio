Pandas: Tablas Inteligentes

```python
import pandas as pd
import numpy as np
```

```python
# Crear desde diccionario: clave=nombre columna, valor=lista de datos
df = pd.DataFrame({
    'producto' : ['Lapto','Mouse', 'Teclado','Monitor'],
    'precio': [2500000,45000,120000,800000],
    'stock':    [15,200,85,30],
    'categoria':['PC','PC','PC','PC']
})

print(df)
print("\nForma:", df.shape)# (4, 4) → 4 filas, 4 columnas
print("Tipos:\n", df.dtypes)
```

Exploración — Los 5 Comandos Siempre

```python
df.head()           # primeros 5 registros
df.tail(3)          # últimos 3
df.info()            # tipos de dato, nulos, memoria
df.describe()        # estadísticas de columnas numéricas
df.isnull().sum()   # cuántos nulos por columna

# Frecuencia de cada categoría
df['categoria'].value_counts()
```

Seleccionar y Filtrar

```python
# Una columna → Series
precios = df['precio']

# Varias columnas → DataFrame
sub = df[['producto' , 'precio']]

# Filtrar filas con condición booleana
caros = df[df['precio'] > 200000]

# Múltiples condiciones
filtro = df[(df['precio'] > 100000) & (df['stock'] > 20)]

# Por posición: iloc[filas, columnas]
# Por posición: iloc[filas, columnas]
df.iloc[0]            # primera fila
df.iloc[1:3]          # filas 1 y 2
df.iloc[:, 1:3]       # todas las filas, columnas 1 y 2

# Por etiqueta: loc
df.loc[df['precio'] == df['precio'].max()]  # fila del más caro
```

Limpieza de Datos

```python
df2 = pd.DataFrame({
    'nombre': ['Ana','Luis','Ana', None,'Marta'],
    'edad':   [25,30,25,28,None],
    'ventas': [1000,2000,1000,1500,1200]
})

# Ver nulos por columna
print(df2.isnull().sum())
# nombre 1  edad 1  ventas 0

# Estrategia 1: eliminar filas con nulos
df_clean = df2.dropna()

# Estrategia 2: rellenar nulos
df2['edad'] = df2['edad'].fillna(df2['edad'].mean())
df2['nombre'] = df2['nombre'].fillna('sin nombre')

# Eliminar duplicados
df2 = df2.drop_duplicates()

# Crear nueva columna
df2['ventas_usd'] = (df2['ventas'] / 4200).round(2)

# Columna condicional
df2['rendimiento'] = np.where(df2['ventas'] >= 1500, 'Alto', 'Bajo')
```

GroupBy — Agrupar y Resumir

```python
# Promedio de ventas por departamento
df.groupby('categoria')['precio'].mean()

# Múltiples estadísticas a la vez
df.groupby('categoria').agg({
    'precio': ['mean','max','min'],
    'stock':  ['sum','count']
})

# Ordenar resultado
df.sort_values('precio', ascending=False)
```

```python
df3 = pd.DataFrame({'nombre':['Carlos','María','Juan','Ana','Pedro','Laura'],
                    'departamento':['Ventas','IT','Ventas','IT','RRHH','Ventas'],
                    'salario':[3200000,5500000,2800000,6000000,3500000,3800000],
                    'años':[3,7,1,10,5,4]})

print(df3)
```

Limpieza Real de Datos
Dataset sucio: {'cliente':['Juan','Ana','Juan',None,'Pedro'], 'compra':[15000,None,15000,8000,22000], 'ciudad':['Cali','Bogotá','Cali','Medellín','Cali']}

Identifica nulos por columna.
Rellena 'compra' nulo con el promedio de las compras conocidas.
Rellena 'cliente' nulo con 'Sin identificar'.
Elimina duplicados.
Crea columna 'compra_usd' dividiendo entre 4200.

```python
df = pd.DataFrame({
    'cliente':['Juan','Ana','Juan',None,'Pedro'],
    'compra': [15000,None,15000,8000,22000],
    'ciudad': ['Cali','Bogotá','Cali','Medellín','Cali']
})

print("Nulos:\n", df.isnull().sum())
# cliente 1  compra 1  ciudad 0

df['compra']  = df['compra'].fillna(df['compra'].mean())  # 15000.0
df['cliente'] = df['cliente'].fillna('Sin identificar')

df = df.drop_duplicates()
print("Filas limpias:", len(df))  # 4

df['compra_usd'] = (df['compra'] / 4200).round(2)
print(df)

```
