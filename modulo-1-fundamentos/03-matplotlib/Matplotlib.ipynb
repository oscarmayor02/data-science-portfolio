Matplotlib: Habla con Gráficas


Un análisis sin visualización es como un mapa sin dibujar. Matplotlib te permite comunicar hallazgos de forma que cualquier persona entienda, incluso sin saber de datos.

Los 4 Tipos Esenciales

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Crear una figura con una grilla de 2x2 gráficos
fig, ax = plt.subplots(2,2, figsize=(14,10))

# Título general de toda la figura
fig.suptitle('Los 4 Tipos de Gráficas Esenciales', fontsize=16)


# ── 1. LÍNEA: para tendencias en el tiempo ──

# Genera 100 valores entre 0 y 12 (eje X)
x = np.linspace(0,12,100)

# Dibuja una línea (simulando ventas)
ax[0,0].plot(x, np.sin(x)*100+200, color='#2E75B6', lw=2)

# Título del gráfico
ax[0,0].set_title('Línea — Tendencias')

# Etiquetas de los ejes
ax[0,0].set_xlabel('Mes')
ax[0,0].set_ylabel('Ventas')

# Muestra una grilla (cuadrícula)
ax[0,0].grid(True, alpha=0.3)


# ── 2. BARRAS: para comparar categorías ──

# Categorías (eje X)
cats = ['Ene','Feb','Mar','Abr','May']

# Valores (eje Y)
vals = [450,620,380,710,590]

# Colores: el valor más alto en rojo, los demás en gris
clrs = ['#C0392B' if v==max(vals) else '#85929E' for v in vals]

# Crear gráfico de barras
ax[0,1].bar(cats, vals, color=clrs)

# Título
ax[0,1].set_title('Barras — Comparación')


# ── 3. HISTOGRAMA: para ver distribuciones ──

# Genera 500 datos con distribución normal (media 175, desviación 8)
datos = np.random.normal(175,8,500)

# Crear histograma con 25 barras
ax[1,0].hist(datos, bins=25, color='#1ABC9C', edgecolor='white')

# Línea vertical mostrando el promedio
ax[1,0].axvline(np.mean(datos), color='red', lw=2, label='Promedio')

# Título
ax[1,0].set_title('Histograma — Distribución')

# Mostrar leyenda
ax[1,0].legend()


# ── 4. DISPERSIÓN: para ver correlaciones ──

# Genera horas de estudio aleatorias
hrs = np.random.uniform(1,10,60)

# Genera notas (relacionadas con horas + ruido)
nota = hrs * 0.5 + np.random.randn(60) * 0.8

# Crear gráfico de dispersión
ax[1,1].scatter(hrs, nota, alpha=0.6, color='#8E44AD')

# Título
ax[1,1].set_title('Dispersión — Correlación')

# Etiquetas de ejes
ax[1,1].set_xlabel('Horas de estudio')
ax[1,1].set_ylabel('Nota')


# Ajusta automáticamente los espacios entre gráficos
plt.tight_layout()

# Muestra todo en pantalla
plt.show()
```

Dashboard de Ventas
Ejercicio

Datos: [450,520,380,610,720,690,580,640,810,750,680,920] (ventas mensuales). Crea 2 gráficas lado a lado: (1) Línea con línea horizontal roja en el promedio. (2) Barras donde los meses sobre el promedio sean verdes y los demás rojos.

```python
import numpy as np                  # Librería para cálculos numéricos (promedio, arrays)
import matplotlib.pyplot as plt     # Librería para hacer gráficas

# Lista con los meses (eje X)
meses = ['Ene','Feb','Mar','Abr','May','Jun',
         'Jul','Ago','Sep','Oct','Nov','Dic']

# Datos de ventas (eje Y)
v = np.array([450,520,380,610,720,690,
              580,640,810,750,680,920])

# Calcular el promedio de las ventas
prom = np.mean(v)


# Crear una figura (ventana) con tamaño ancho
plt.figure(figsize=(14,5))


# ─────────────────────────────
# GRÁFICA 1: LÍNEA
# ─────────────────────────────
plt.subplot(1,2,1)
# (1 fila, 2 columnas, posición 1 → izquierda)

# Dibujar línea de ventas
plt.plot(meses, v, marker='o')
# marker='o' pone punticos en cada dato

# Dibujar línea horizontal del promedio
plt.axhline(prom, color='red')
# axhline = línea horizontal

# Título del gráfico
plt.title('Ventas (Línea)')


# ─────────────────────────────
# GRÁFICA 2: BARRAS
# ─────────────────────────────
plt.subplot(1,2,2)
# (1 fila, 2 columnas, posición 2 → derecha)

# Crear colores según el promedio
colores = ['green' if x > prom else 'red' for x in v]
# Si la venta es mayor al promedio → verde
# Si no → rojo

# Dibujar barras
plt.bar(meses, v, color=colores)

# Dibujar línea del promedio también aquí
plt.axhline(prom, color='black')

# Título del gráfico
plt.title('Ventas (Barras)')


# Ajustar espacios para que no se encimen
plt.tight_layout()

# Mostrar todo en pantalla
plt.show()
```

EJERCICIO

Tienes estos datos de temperatura semanal:

[22, 24, 19, 23, 27, 26, 21]

Haz 2 gráficas lado a lado:

Gráfica de línea
Mostrar la temperatura por día
Dibujar una línea horizontal con el promedio
Gráfica de barras
Barras verdes si la temperatura está por encima del promedio
Barras rojas si está por debajo

```python
# Días de la semana (eje X)
dias = ['Lun','Mar','Mié','Jue','Vie','Sáb','Dom']

# Temperaturas (eje Y)
temp = np.array([22, 24, 19, 23, 27, 26, 21])

# Calcular el promedio
prom = np.mean(temp)

# Crear figura
plt.figure(figsize=(12,5))

# ─────────────────────────────
# GRÁFICA 1: LÍNEA
# ─────────────────────────────
plt.subplot(1,2,1)

# Línea de temperatura
plt.plot(dias, temp, marker='o')

# Línea del promedio
plt.axhline(prom, color='red')

# Título
plt.title('Temperatura semanal (Línea)')


# ─────────────────────────────
# GRÁFICA 2: BARRAS
# ─────────────────────────────
plt.subplot(1,2,2)

# Colores según el promedio
colores = ['green' if x > prom else 'red' for x in temp]

# Barras
plt.bar(dias, temp, color=colores)

# Línea del promedio
plt.axhline(prom, color='black')

# Título
plt.title('Temperatura vs promedio')


# Ajustar espacios
plt.tight_layout()

# Mostrar
plt.show()
```
